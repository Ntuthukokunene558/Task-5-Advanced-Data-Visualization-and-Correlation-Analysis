# Task 5-Advanced Data Visualization and Correlation Analysis

```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the dataset
spectacles_data = pd.read_csv('Dots Potential Customer Survey Data - Sheet1.csv')
```

```python
spectacles_data.head()
```

## Advanced visualization techniques provide deeper insights into the data.

```python
# Heatmap of correlations
correlation_matrix = spectacles_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Explore the correlation between variables
plt.figure(figsize=(12, 8))
sns.heatmap(spectacles_data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
f,axes = plt.subplots(1,1,figsize = (12,12))
sns.heatmap(spectacles_data.corr(),annot = True, ax = axes)
```

```python
# Correlation Analysis
target_correlation = correlation_matrix['Wear Specs'].drop('Wear Specs')
strong_correlation_threshold = 0.5
strong_correlation_features = target_correlation[abs(target_correlation) > strong_correlation_threshold]
print("Features strongly correlated with Wear Specs:")
print(strong_correlation_features)
```

```python
# Correlation Analysis
correlation_matrix = spectacles_data.corr()
```

```python
# Correlation with target variable (Wear Specs)
correlation_with_target = correlation_matrix['Wear Specs'].sort_values(ascending=False)
```

```python
print("Correlation with Wear Specs:")
print(correlation_with_target)
```

```python
# Creating the heatmap
sns.heatmap(correlation_matrix)
```

```python
k = 10  # number of variables for heatmap
cols = spectacles_data.corr().nlargest(k, 'Wear Specs')['Wear Specs'].index
cm = spectacles_data[cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, cmap='coolwarm')  # Replace 'viridis' with the desired colormap
plt.show()
```
